# Baseline Function
used to find the distance between two dishes in the array

input a list of dish indexes and return the distance between them in an array formatted [East, North, Up]

```python
def baseline(east, north, up=0, dist=14.6):
    A = np.array([east, north, up]) * dist
    return A
```

testing `baseline()`

```python
baseline(1,1,0)
```

**Next, make a function that will generate an array of locations**

The `square()` function is working well 

```python
def square(length):
    a = []
    for i in range(1,length+1):
        for j in range(1,length+1):
            a.append([i,j,0])
    return a
```

**still a work in progress**

```python
def triangle(length):
    a = []
    mid = length//2
    height = int(np.sqrt(3)/2*length)
    for j in range(0,height):
        for i in range(-mid,mid):
            if i <= (height/mid)*j + height and i <= -(height/mid)*j + height:
                a.append([i,j,0])
    return a
```

```python
def generateLocations(length, shape="square"): # figure out what will be good arguments 
    a = []
    if shape == "square":
        return square(length)
    elif shape == "triangle":
        return triangle(length)
    return a
```

```python
def plotShape(side, shape="square"):
    coords = generateLocations(side, shape)    
    x = [coords[i][0] for i in range(0,len(coords))]
    y = [coords[i][1] for i in range(0, len(coords))]
    plt.scatter(x,y)
    plt.xlabel("East")
    plt.ylabel("North")
    plt.title("Locations of points")
    plt.show()
```

```python
plotShape(10,"square")
```
